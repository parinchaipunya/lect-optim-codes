```julia
using JuMP, SCIP
```

```julia
R = [.175; .055];
Sigma = [.067 -.005; -.005 .013];

P = zeros(50,2);
for i in 1:50
    eta = .02*i;
    port = Model(SCIP.Optimizer);
    @variable(port, p[1:2] >= 0);
    @variable(port, obj);
    @constraint(port, sum(p[i] for i in 1:2) == 1);
    @constraint(port, obj >= sum(Sigma[i,j]*p[i]*p[j] for i in 1:2 for j in 1:2) 
        - eta*sum(R[i]*p[i] for i in 1:2))
    @objective(port, Min, obj );
    optimize!(port)
    P[i,:] = value.(p)
end
```
