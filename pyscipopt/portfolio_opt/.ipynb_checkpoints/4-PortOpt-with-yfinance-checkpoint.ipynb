# Portfolio Optimization with `yfinance`

Here we can screen for stock data from [Yahoo Finance](https://finance.yahoo.com/) using the package `yfinance`, as shown in the following example.

```python
import numpy as np
from pyscipopt import Model, quicksum
import matplotlib.pyplot as plt

def PortOpt(R,Sigma):

    Etas = np.arange(0.02,1.01,0.02)
    cases = np.shape(Etas)[0]
    P = cases * [None]
    Ret = cases * [None]
    Var = cases * [None]
    n = len(R)
    
    for e in range(cases):
        eta = Etas[e]
    
        # Create the model
        port = Model("Portfolio optimization for two assets")
    
        # Add variables and constraints
        p = [None for i in range(n)]
        for i in range(n):
            p[i] = port.addVar(vtype='C', lb=0, ub=None, name=f"p_{i}")
        port.addCons(quicksum(p[i] for i in range(n)) == 1, name="budget")
    
        # The objective is Nonlinear. We need the following TRICK.
        obj = port.addVar(vtype='C', lb=None, ub=None, name='objective value')
        port.addCons(obj >= quicksum(Sigma[i][j]*p[i]*p[j] for i in range(n) for j in range(n)) 
                     - eta*quicksum(R[i]*p[i] for i in range(n)))
        port.setObjective(obj, sense="minimize")
    
        # Solve
        port.hideOutput()
        port.optimize()
    
        # Collect solutions
        SOL = port.getBestSol()
        Port = np.array(n*[None])
        for i in range(n):
            Port[i] = SOL[p[i]]
        P[e] = Port
        Ret[e] = np.dot(R,P[e])
        Var[e] = np.dot(P[e],np.dot(Sigma,P[e]))
    
    Ports = np.array(P)
    plt.plot(Var,Ret)
    plt.show()

    return Ports
```

```python
import yfinance as yf
import pandas as pd
```

```python
thai_stocks = ['BBL.BK', 'KBANK.BK', 'SCB.BK', 'PTT.BK', 'PTTEP.BK',
               'ADVANC.BK', 'AOT.BK', 'TRUE.BK', 'CPALL.BK', 'CPF.BK',
               'GULF.BK', 'BH.BK', 'BEM.BK', 'CPN.BK', 'HMPRO.BK',
               'BDMS.BK', 'MINT.BK', 'EGCO.BK', 'RATCH.BK', 'BGRIM.BK']

start_date = '2023-10-01'
end_date = '2024-10-31'

# Initialize an empty DataFrame to store the results
all_stock_data = pd.DataFrame()

# Loop through each stock and get monthly close prices
for stock in thai_stocks:
    data = yf.download(stock, start=start_date, end=end_date, interval='1mo')
    data = data[['Close']].rename(columns={'Close': stock})
    all_stock_data = pd.concat([all_stock_data, data], axis=1)

# Reset index and format
all_stock_data.index = all_stock_data.index.strftime('%B-%y')
all_stock_data.reset_index(inplace=True)
all_stock_data.rename(columns={'index': 'Month'}, inplace=True)
all_stock_data.columns = ['Month'] + thai_stocks  # Rename columns to only "Month" and stock names
all_stock_data.set_index('Month', inplace=True)
```

```python
all_stock_data
```

```python
# Compute returns using pct_change()
# returns = all_stock_data.pct_change().dropna()
returns = all_stock_data.pct_change()

# # Check for NaN or infinite values and handle them
# returns = returns.replace([np.inf, -np.inf], np.nan).dropna()

# Calculate mean and covariance
R = returns.mean().values
Sigma = returns.cov().values
```

```python
print(f"Return vector:\nR =\n{R}")
print(f"Covariance matrix:\n Sigma =\n{Sigma}")
```

```python
Ports = PortOpt(R,Sigma)
```
